```python
def con_dis_split(df, thr=36):
    con = []
    dis = []
    for i in range(df.shape[1]):
        if df.iloc[:,i].unique().size >=thr: 
            con.append(i)
        else: 
            dis.append(i)
    return df.iloc[:,con],df.iloc[:, dis]
```

```python
customer_con,customer_dis = con_dis_split(customer_num)

display(customer_con.head())
customer_dis.head()
```

```python
def barplot_df(df):
    fig,ax = plt.subplots(1,df.shape[1],figsize=(16,16))
    for i in range(0,df.shape[1]):
        unique_values = df.iloc[:,i].unique().size
        if (unique_values <= 6):
            sns.countplot(x=df.iloc[:,i], order=df.iloc[:,i].value_counts().index,ax=ax[i])
        else:
            sns.countplot(y=df.iloc[:,i], order=df.iloc[:,i].value_counts().index,ax=ax[i])
    return fig
```

```python
def histplot_df(df):
    fig,ax = plt.subplots(1,df.shape[1],figsize=(16,16))
    for i in range(0,df.shape[1]):
        sns.histplot(x=df.iloc[:,i],ax=ax[i])
    return fig
```

```python
def clean_outliers(df,col):
    max_thr = (int(df[col].quantile([0.75]))-int(df[col].quantile([0.25])))*1.5+int(df[col].quantile([0.75]))
    df[col]=df[col][df[col]<=max_thr]
    return df[col]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=31)
```

```python
X_train_num = X_train.select_dtypes(include = np.number)
X_test_num = X_test.select_dtypes(include = np.number)
X_train_cat = X_train.select_dtypes(include = 'object')
X_test_cat = X_test.select_dtypes(include = 'object')
```

```python
encoder = OneHotEncoder(drop='first')
encoder.fit(X_train_cat)
encoded = encoder.transform(X_train_cat).toarray()
X_train_coded = pd.DataFrame(encoded,columns=encoder.get_feature_names_out(),index=X_train_cat.index)
X_train_coded.head()
```

```python
encoded_2 = encoder.transform(X_test_cat).toarray()
X_test_coded = pd.DataFrame(encoded_2,columns=encoder.get_feature_names_out(),index=X_test_cat.index)
X_test_coded.head()
```

```python
X_test_pt_np = pt.transform(X_test_num)
X_test_pt = pd.DataFrame(X_test_pt_np,columns=X_test_num.columns,index=X_test_num.index)
X_test_pt.head()
```

```python
X_train_new =pd.concat([X_train_pt,X_train_coded],axis=1)
X_test_new = pd.concat([X_test_pt,X_test_coded],axis=1)
```

```python
def linear_model(X1,X2,y1,y2):
    from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
    from sklearn import linear_model
    lm = linear_model.LinearRegression()
    lm.fit(X1,y1)
    y_pred = lm.predict(X2)
    MAE = mean_absolute_error(y2,y_pred)
    MSE = mean_squared_error(y2,y_pred)
    RMSE = np.sqrt(mean_squared_error(y2,y_pred))
    MAPE = mean_absolute_percentage_error(y2,y_pred)
    R2 = r2_score(y2,y_pred)
    
    dic = {"Error_metric":["MAE","MSE","RMSE","MAPE","R2"],"Value":[round(MAE,2),round(MSE,2),round(RMSE,2),round(MAPE,2),round(R2,2)]}
    value = pd.DataFrame(data = dic)
    return value
```

```python
linear_model(X_train_new_scaled_df,X_test_new_scaled_df,y_train,y_test)
```

```python
def model_training(X1,X2,y1,y2):
    from sklearn.neighbors import KNeighborsRegressor
    from sklearn.metrics import r2_score
    from sklearn import linear_model
    from sklearn.neural_network import MLPRegressor
    
    lm = linear_model.LinearRegression()
    lm.fit(X1,y1)
    y_pred = lm.predict(X2)
    R2 = r2_score(y2,y_pred)
    
    neigh = KNeighborsRegressor(n_neighbors=6)
    neigh.fit(X1,y1)
    y_pred_2=neigh.predict(X2)
    R2_KN = r2_score(y2,y_pred_2)
    
    regr = MLPRegressor(random_state=1, max_iter=10000).fit(X1,y1)
    regr.predict(X2)
    score = regr.score(X2,y2)
    
    value_list = [R2,R2_KN,score]
    df = pd.DataFrame(data = value_list, index=['Linear Regression','KNeighborsRegressor','MLPRegressor'],columns=['R2 score'])
    
    return df
```

```python
model_training(X_train_new_scaled_df,X_test_new_scaled_df,y_train,y_test)
```
